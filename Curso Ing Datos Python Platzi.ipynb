# ETL = Extract Transform Load

### Extract: 
Es el proceso de lectura de datos de diversas fuentes
* Base de datos
* CRM
* Archivos CSV
* Datasets públicos

### Transform: 
En este momento cuando nosotros tenemos que transformar los datos, tenemos que identificar datos faltantes o datos erróneos o una edad negativa. En esta etapa donde tenemos que identificar todos los problemas y solucionarlos.
* Limpieza
* Estructurado
* Enriquecimiento.

### Load:
Una vez transformados debemos insertarlos en el data warehouse
Depende del tipo de solución que se haya escogido

### Series
- Vector unidimensional con un índice
- Datos homogéneos
- Se pueden crear a partir de cualquier secuencia
- Óptimo si se usan dicionarios 

```
series = pd.Series([1,2,3])
#mejor: 
series = pd.Series({'a':1,'b':2,'c':3})
```





````
import pandas as pd
series_test = pd.Series([100,200,300])
series_test
````

````
#con índice
series_test2 = pd.Series({1999:48, 2000:65, 2001:89})
series_test2
````

### DataFrames
- Tabla con filas y columnas etiquetadas
- Pueden ser arrays bidimensionales de Numpy, listas de listas, de tuplas, un diccionario u otros dataframe
- Sus llaves se convierten en el índice de sus columnas

````
frame_test = pd.DataFrame({1999:[78,45,56],
                           2000:[45,78,59],
                           1987:[52,89,14]})
frame_test
````

````
frame_test2 = pd.DataFrame([[78,45,56],
                           [45,78,59],
                           [52,89,14]], columns=[1999,2000,1987])
frame_test2
````

### Índices y selección
 
#### - Dictionary like
```
# Regresa un DataSeries:
df['col1'] 
# Regresa un DataFrame:
df['col1', 'col3'] 
```

#### - Numpy like
```
# Regresa fila:
df.iloc[:] 
# Regresa fila, columna:
df.iloc[:,:] 
```

#### - Label based
```
# Regresa fila:
df.loc[:] 
# Regresa fila, columna:
df.loc[:,:] 
```
